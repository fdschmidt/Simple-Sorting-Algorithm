# Backup: Simple Sorting Algorithm

##### Loading the Libraries

```python
# Import Libraries
import random
import matplotlib.pyplot as plt
import numpy as np
```

### Functions

###### RandomList(length) 
This function generates a list of length len with random numbers between 0 and 1

```python
def RandomList(length):
    # Initialize result
    randomList = []
    
    # Fill list with random numbers
    for i in range(0,length):
        randomList.append(random.random())
        
    # Return result
    return randomList
```

###### RandomizeList() 
This function randomizes a list of arbitrary length

```python
def RandomizeList(sortedList):
    # Initialize result
    randomizedList = []
    
    while (len(sortedList) > 0):
        # Generate a random index for the row
        randomIndex = random.randint(0, len(sortedList)-1)
        
        # Pop the element in the initial row corresponding to
        # randomIndex and append it to the randomized row 
        randomizedList.append(sortedList.pop(randomIndex))
    
    # Return result
    return randomizedList
```

###### RandomMatrix(X, Y)
This function generates a matrix with X columns and Y rows filled with random numbers. All rows contain exactly the same numbers but the order of the elements is randomized.

```python
def RandomMatrix(X, Y):
    # Initialize random matrix
    randomMatrix = []
    
    # Generate a random list of length X
    randomList = RandomList(X)
    
    # Randomize the list Y times and append each randomized list 
    # as a new row to the matrix
    for i in range(0,Y):
        # Copy list to avoid memory conflict
        tmpList = randomList[:]
        
        # Randomize list
        randomizedList = RandomizeList(tmpList)
        
        # Append to matrix
        randomMatrix.append(randomizedList)
    
    # Return result
    return randomMatrix
```

###### SortMatrix()
This function sorts the rows of a matrix from small to large numbers.

```python
def SortMatrix(unsortedMatrix):

    # Initialize result
    sortedMatrix = []

    # Loop over original matrix
    for i in range(0,len(unsortedMatrix)):
    
        # Append new row
        sortedMatrix.append([])
    
        # Loop over every element in each row of the original matrix
        for j in range(0,len(unsortedMatrix[0])):
        
            # Case 1: Submatrix is empty
            if len(sortedMatrix[i]) == 0:
                index = 0
                sortedMatrix[i].insert(index, unsortedMatrix[i][j])
            # Case 2: Submatrix already contains elements
            else:
                # Case 2.1: New element is smaller than currently smallest element
                if (unsortedMatrix[i][j] <= sortedMatrix[i][0]):
                    index = 0
                    sortedMatrix[i].insert(index, unsortedMatrix[i][j])
                # Case 2.2: New element is larger than currently largest element
                elif (unsortedMatrix[i][j] >= sortedMatrix[i][-1]):
                    index = len(sortedMatrix[i])
                    sortedMatrix[i].insert(index, unsortedMatrix[i][j])
                # Case 2.3: New element is somwhere in between and needs to be sorted
                else:
                    for k in range(0,len(sortedMatrix[i])):
                        # Find corresponding index to corrext insertion point
                        if unsortedMatrix[i][j] >= sortedMatrix[i][k]:
                            continue
                        else:
                            # Insertion point found, set index
                            index = k
                            # Insert into sorted matrix
                            sortedMatrix[i].insert(index, unsortedMatrix[i][j])
                            break
                            
    # Return result
    return sortedMatrix
```

### Main

```python
# Generate a 500x500 matrix of random numbers between 0 and 1 where 
# each row contains the same numbers (but shuffled)
unsortedMatrix = RandomMatrix(500,500)

# Sort each of row of the matrix from smallest to larges number
sortedMatrix = SortMatrix(unsortedMatrix)
```

### Plot 

###### Visualize unsorted matrix

```python
# Generate cartesian grid
x = np.arange(0,1,0.002)
y = np.arange(0,1,0.002)


# Set height and width of figure
width = 8
height = 8

# Initialize figure and subplot
fig = plt.figure(figsize=(width, height))
pl = fig.add_subplot(111)

# Set limits
pl.set_xlim(0, 1.0)
pl.set_ylim(0, 1.0)

# Set lables
pl.set_title("Original", fontsize=16)
pl.set_xlabel("x", fontsize=16)
pl.set_ylabel("y", fontsize=16)

# Generate plot
plot = pl.pcolormesh(x, y, unsortedMatrix, cmap="viridis")

# Display figure
plt.show()
```

###### Visualize unsorted and sorted matrix

```python
# Set height and width of figure
width = 15
height = 8

# Initialize figure and subplot
fig = plt.figure(figsize=(width, height))


# Generate subplot 1
# ------------------
pl1 = fig.add_subplot(121)

# Set limits
pl1.set_xlim(0, 1.0)
pl1.set_ylim(0, 1.0)

# Set lables
pl1.set_title("Original Matrix", fontsize=16)
pl1.set_xlabel("x", fontsize=16)
pl1.set_ylabel("y", fontsize=16)

# Generate plot
plot = pl1.pcolormesh(x, y, unsortedMatrix, cmap="viridis")


# Generate subplot 2
# ------------------
pl2 = fig.add_subplot(122)

# Set limits
pl2.set_xlim(0, 1.0)
pl2.set_ylim(0, 1.0)

# Set lables
pl2.set_title("Sorted Matrix", fontsize=16)
pl2.set_xlabel("x", fontsize=16)
pl2.set_ylabel("y", fontsize=16)

# Generate plot
plot = pl2.pcolormesh(x, y, sortedMatrix, cmap="viridis")


# Display figure
plt.tight_layout()
plt.show()
```
